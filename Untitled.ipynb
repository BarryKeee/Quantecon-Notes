Steps for LQ Approximation

1. Solve for the steady state
2. Identify the endogenous and exogenous states, and the choice variable
3. Redefine the utility as a function of (endogenous and exogenous) state variables and choice variables (usually use constraints)
4. Approximate the utility function around the steady state using 2nd order Taylor approximation
5. Use value function iteration to find the optimal value function
